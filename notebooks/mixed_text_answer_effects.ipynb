```python
merged_h, merged_g = M.load_merged_data(
    hunters_path="../data/merged_hunters.csv",
    gatherers_path="../data/merged_gatherers.csv",
)
```

```python
models_h_sep = M.fit_no_slopes_for_all_answers(merged_h, M.TERMS, separated=True)
models_g_sep = M.fit_no_slopes_for_all_answers(merged_g, M.TERMS, separated=True)

M.plot_fixed_effects(M.TERMS, models_h_sep, h_or_g="hunters", separated_label="answ_separated")
M.plot_fixed_effects(M.TERMS, models_g_sep, h_or_g="gatherers", separated_label="answ_separated")
```

```python
models_h_all = M.fit_no_slopes_for_all_answers(merged_h, M.TERMS, separated=False)
models_g_all = M.fit_no_slopes_for_all_answers(merged_g, M.TERMS, separated=False)

M.plot_fixed_effects(M.TERMS, models_h_all, h_or_g="hunters", separated_label="all_answ")
M.plot_fixed_effects(M.TERMS, models_g_all, h_or_g="gatherers", separated_label="all_answ")
```

```python
models_h_sloped = {}
models_g_sloped = {}


#not enough observations if both values are true
sep = False
corr = True

for label in C.ANSWER_LABELS:
    m_h = M.build_model_with_slopes(
        merged_h,
        answer_label=label,
        terms=M.TERMS,
        separated=sep,
        correlated_slopes=corr,
    )
    models_h_sloped[label] = m_h
    M.plot_participant_effects(
        m_h, f"Answer {label}",
        M.TERMS, h_or_g="hunters",
        separated=sep,
        correlated_slopes=corr
        )

    m_g = M.build_model_with_slopes(
        merged_g,
        answer_label=label,
        terms=M.TERMS,
        separated=sep,
        correlated_slopes=corr,
    )
    models_g_sloped[label] = m_g
    M.plot_participant_effects(
        m_g, f"Answer {label}",
        M.TERMS, h_or_g="gatherers",
        separated=sep,
        correlated_slopes=corr
        )


sep_label = "separated" if sep else "all_trials"
corr_label = "correlated" if corr else "uncorrelated"

M.plot_text_to_answer_significance_heatmap(
    models=models_h_sloped,
    terms=M.TERMS,
    title="hunters — text→answer fixed effects ({}, {})".format(sep_label, corr_label),
    alpha=0.05,
    adjust="holm",
    save_path=os.path.join(
    "..",
    "reports",
    "plots",
    "texts_to_answers",
    "significance_heatmaps",
    "{}__{}__hunters.png".format(sep_label, corr_label),
    )
)

M.plot_text_to_answer_significance_heatmap(
    models=models_g_sloped,
    terms=M.TERMS,
    title="gatherers — text→answer fixed effects ({}, {})".format(sep_label, corr_label),
    alpha=0.05,
    adjust="holm",
    save_path=os.path.join(
    "..",
    "reports",
    "plots",
    "texts_to_answers",
    "significance_heatmaps",
    "{}__{}__gatherers.png".format(sep_label, corr_label),
    )
)
```

```python
all_participants = pd.concat([merged_h, merged_g], ignore_index=True)
models_all_sloped = {}

sep = False
corr = True

for label in C.ANSWER_LABELS:
    m_all = M.build_model_with_slopes(
        all_participants,
        answer_label=label,
        terms=M.TERMS,
        separated=sep,
        correlated_slopes=corr,
    )
    models_all_sloped[label] = m_all
    M.plot_participant_effects(
        m_all, f"Answer {label}",
        M.TERMS, h_or_g="all_participants",
        separated=sep,
        correlated_slopes=corr
        )


sep_label = "separated" if sep else "all_trials"
corr_label = "correlated" if corr else "uncorrelated"

M.plot_text_to_answer_significance_heatmap(
    models=models_all_sloped,
    terms=M.TERMS,
    title="all_participants — text→answer fixed effects ({}, {})".format(sep_label, corr_label),
    alpha=0.05,
    adjust="holm",
    save_path=os.path.join(
    "..",
    "reports",
    "plots",
    "texts_to_answers",
    "significance_heatmaps",
    "{}__{}__all_participants.png".format(sep_label, corr_label),
    )
)
```
